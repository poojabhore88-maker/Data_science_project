```python
plt.figure(figsize=(14,6))
top_cuisines = df["Cuisines"].value_counts().head(10)
sns.barplot(x=top_cuisines.index, y=top_cuisines.values, palette="coolwarm")
plt.title("Top 10 Most Common Cuisines")
plt.xticks(rotation=80)
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(12,5))
country_counts = df["Country Code"].value_counts()
sns.barplot(x=country_counts.index.astype(str),
            y=country_counts.values,
            palette="mako")
plt.title("Country Code Distribution")
plt.xlabel("Country Code")
plt.ylabel("Frequency")
plt.show()
```

```python
import folium
from folium.plugins import HeatMap
```

```python
df = df[(df['Latitude'].notnull()) & (df['Longitude'].notnull())]

# Convert to numeric if needed
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')

df = df.dropna(subset=['Latitude','Longitude'])

print("\nValid coordinates remaining:", len(df))
```

```python
center_lat = df['Latitude'].mean()
center_lon = df['Longitude'].mean()

base_map = folium.Map(location=[center_lat, center_lon], zoom_start=5)
for _, row in df.sample(min(300, len(df))).iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=2,
        popup=row['Restaurant Name'],
        color="blue",
        fill=True,
        fill_opacity=0.7
    ).add_to(base_map)

base_map
```

```python
heat_data = df[['Latitude', 'Longitude']].values.tolist()

heat_map = folium.Map(location=[center_lat, center_lon], zoom_start=5)
HeatMap(heat_data, radius=10).add_to(heat_map)

heat_map
```

```python
plt.figure(figsize=(14,6))
top_city_count = df['City'].value_counts().head(10)
sns.barplot(x=top_city_count.index, y=top_city_count.values, palette="plasma")
plt.xticks(rotation=45)
plt.title("Top 10 Cities with Highest Restaurant Density")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='Longitude', y='Latitude', hue='Aggregate rating',
                palette="Spectral", s=30)
plt.title("Restaurant Ratings Mapped to Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
